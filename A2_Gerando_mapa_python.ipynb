```python
import pandas as pd

# Essa biblioteca fornece ferramentas para análise e manipulação de dados.
# DADOS tabulares como planilha de EXCEL ou uma tabela SQL
# DADOS ordenados de modo temporal ou não
# MATRIZES 
```

```python
# PEGANDO O NOME DE TODOS OS MUNICIPIOS DO NOSSO GEODATAFRAME
municipio = pd.unique(censo_sp_2010.name_muni)
```

```python
# MOSTRA O NÚMERO DE MUNICIPIOS
print(len(municipio))
```

```python
print(municipio)
```

```python
# BIBLIOTECA USADO PARA CAUSAR UMA PAUSA NA EXECUÇÃO DO CODIGO COM TIME.SLEE
import time
```

```python
for mun in municipio:
  print(mun)
  time.sleep(10)
```

```python
#ARMAZENANDO NOME DE MUNICIPIOS QUE COMEÇAM COM 'ARA
nomes =[]

for mun in municipio:
  if mun.startswith('Mogi'):
    nomes.append(mun)

nomes
```

```python
#OUTRO TIPO DE FOR (ESTUDAR)
nomes=[mun for mun in municipio if mun.startswith('Ara')]
nomes
```

```python
#SELECIONANDO APENAS OS SETORES CENSITÁRIOS DE ARARAQUARA E ARMAZENANDO EM UMA NOVA VARIÁVEL

censo_mogi_2010= censo_sp_2010[censo_sp_2010['name_muni']=='Mogi Das Cruzes']

censo_mogi_2010
```

```python
# VISUALIZANDO A GEOMETRIA DOS DADOS SELECIONANDOS

censo_mogi_2010.plot()
censo_mogi_2010.head()
```
